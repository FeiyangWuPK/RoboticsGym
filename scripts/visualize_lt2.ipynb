```python
import numpy as np
import matplotlib.pyplot as plt
```

## Ant-v4 LT2

```python
# Load your numpy data
from matplotlib.pyplot import step


teacher_data = np.load('/home/feiyang/Documents/Repositories/RoboticsGym/logs/ICML2024 Guided Learning MuJoCo RL/Mujoco RL Ant-v4 0.5/teacher/evaluations.npz')['results']
student_data = np.load('/home/feiyang/Documents/Repositories/RoboticsGym/logs/ICML2024 Guided Learning MuJoCo RL/Mujoco RL Ant-v4 0.5/student/evaluations.npz')['results']

# Calculate the mean and standard deviation for each row
teacher_means = np.mean(teacher_data, axis=1)
teacher_std_devs = np.std(teacher_data, axis=1)

student_means = np.mean(student_data, axis=1)
student_std_devs = np.std(student_data, axis=1)

# Plotting
plt.figure()
plt.plot(teacher_means, label='Teacher')
plt.fill_between(range(len(teacher_means)), teacher_means - teacher_std_devs, teacher_means + teacher_std_devs, alpha=0.3)
plt.plot(student_means, label='Student')
plt.fill_between(range(len(student_means)), student_means - student_std_devs, student_means + student_std_devs, alpha=0.3)
plt.xlabel('Env steps (in millions)')
plt.ylabel('Episodic reward')

# Setting x-axis to display a scale from 0 to 1
plt.xlim(0, len(teacher_means)-1)
num_ticks = 5  # You can change this for more or fewer ticks
tick_locations = np.linspace(0, len(teacher_means)-1, num_ticks)
tick_labels = np.linspace(0, 1, num_ticks)
plt.xticks(tick_locations, tick_labels)

plt.title('Ant-v4')
plt.legend()
plt.savefig('ant_rl.pdf')
```
